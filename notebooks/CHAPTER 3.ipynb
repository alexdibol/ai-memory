##0.CONTEXT

**Introduction — Temporal Intelligence Under Governance: Memory, Belief, Interference, and Decision Traceability**  
Temporal intelligence is often marketed as a capacity upgrade: agents can “remember more,” keep longer chat histories, and operate across larger context windows. In practice, temporal intelligence is not a quantity problem. It is a control problem. Once an agent accumulates a long history—market data, research notes, filings, chat transcripts, internal memos—the central challenge becomes evidence selection under time, uncertainty, and interference. An agent that cannot reliably decide what is relevant now, what has become stale, what is redundant, and what is adversarial does not become smarter with more memory; it becomes more fragile. In finance, the analogy is exact: more data does not guarantee better inference when regimes shift, distributions exhibit heavy tails, and execution constraints transform theoretical signals into costly churn. Chapter 3 is built to make that reality tangible, measurable, and auditable.

The chapter’s premise is that temporal intelligence requires three components working in concert. First, a **latent state model** that represents the evolving environment. In markets, “state” is not directly observable; it is inferred from noisy prices, volumes, and macro signals. Here, the state is a latent regime process modeled as a Markov chain, and the agent’s job is to infer it sequentially. Second, an **episodic memory system** that stores time-stamped evidence, but does not naïvely surface it. Episodic memory must be governed by decay, novelty controls, and contradiction handling; otherwise it becomes a high-dimensional interference field that amplifies false consensus and hard negatives. Third, a **policy and execution layer** that converts belief and evidence into actions under realistic constraints. In finance, action is never free: turnover generates transaction costs and convex impact; volatility targeting and leverage caps impose feasibility. Temporal intelligence must therefore be evaluated end-to-end as a closed-loop system, not as a retrieval demo or a forecasting curve.

The laboratory is synthetic by design, but it is not simplistic. The synthetic market is engineered to reflect the structural features that break naïve models. Returns are heavy-tailed (Student-t), regime-dependent, and punctuated by jumps. Spreads widen under stress, tying market turbulence to execution cost. This matters because in real markets the penalty for being wrong is not symmetric; during stress, liquidity vanishes and costs explode. A model that looks acceptable in quiet periods can become catastrophically fragile when volatility rises and discontinuities appear. By building these features into the environment, the notebook creates a world where temporal inference and governance controls are not optional decorations—they are the conditions for survival.

The latent regime is the ground truth that the agent cannot see. Instead, the agent observes returns and must maintain a posterior distribution over regimes. The notebook uses a particle filter rather than a simple classifier because finance-grade temporal reasoning is inherently uncertain. A point estimate of regime is brittle; a posterior distribution is an operational object. It allows the agent to act conservatively when uncertainty is high and to detect when the model is mismatched. Effective sample size (ESS) and resampling are not mathematical footnotes; they are diagnostics of epistemic health. When ESS collapses, the posterior is dominated by a few particles, and the agent should treat its belief as fragile. In professional risk practice, this is the moment to apply controls: reduce leverage, tighten turnover, or require review. The notebook complements Bayesian filtering with a CUSUM alarm on standardized residuals. This is a model-risk control: even a well-designed filter can be wrong if the environment changes. CUSUM provides a fast, interpretable signal that the observation model is accumulating systematic error. It turns “I have a feeling something changed” into an auditable alarm trace.

Now consider memory. In most discussions, memory is imagined as a storage mechanism: a vector database of embeddings, a longer context window, or an external store. But storage is not the hard part. Selection is the hard part. Once you have thousands of memory items, retrieval becomes a nearest-neighbor contest in a high-dimensional space. Chapter 2 showed that contest is unstable under scale: hubness, near-duplicates, and hard negatives can dominate rankings. Chapter 3 extends that result into time. Evidence is not just high-dimensional; it is time-stamped and regime-coupled. Some evidence becomes stale. Some evidence repeats. Some evidence is adversarially aligned. A temporal agent must therefore implement **governed retrieval** rather than similarity-only retrieval.

This notebook’s episodic memory is governed by three explicit controls. **Temporal decay** applies a half-life to older items, encoding the basic principle that time erodes relevance. This is not a heuristic; it is an epistemic posture. In markets, structural relations drift, narratives invert, and regime-dependent effects flip sign. Treating old evidence as equally relevant is a recipe for anchoring errors. **Novelty gating** penalizes redundancy and hubness. Redundant evidence is especially dangerous in finance because repetition masquerades as confirmation. Ten copies of the same claim are not ten independent signals; they are one claim replicated. Without novelty controls, retrieval will overweight redundant clusters and produce the illusion of strong evidence. **Contradiction gating** penalizes evidence whose implied direction conflicts with the agent’s belief-weighted expected drift. This is not a prohibition; it is a control against whipsaw. In high-noise regimes, raw sentiment bursts can force the agent to rotate aggressively unless contradiction is handled explicitly. Contradiction gating therefore stabilizes decision-making by requiring stronger justification for evidence that fights the current belief state.

A critical constraint is the **retrieval budget**, which is the notebook’s proxy for a context window. Even if memory stores tens of thousands of items, an agent cannot condition on everything. It must choose a bounded set. The retrieval budget makes tradeoffs explicit: higher budgets increase coverage but also increase interference risk and computational burden; lower budgets improve stability but may omit crucial evidence. In production, this budget is shaped by latency, cost, and safety. In finance, it is shaped by reviewability: a human supervisor cannot audit an agent that cites hundreds of snippets. The notebook forces the student to treat the budget as a governance lever, not as an afterthought.

Belief and memory then feed a policy. The policy is intentionally not optimized; it is mechanistic and interpretable. It combines belief-driven expected drift with retrieved evidence polarity through bounded nonlinearities. This is an important methodological shift: the notebook is not teaching “alpha hunting.” It is teaching mechanism literacy: how state inference and memory selection produce actions under constraints. Those constraints are finance-realistic: volatility targeting scales position inversely with estimated risk; leverage caps prevent unrealistic exposures; turnover caps prevent pathological churn; convex impact converts aggressive trading into nonlinear costs. This is the professional point: strategy viability is not determined by signal alone. Viability is determined by the interaction of signal, state, memory, and execution surfaces.

The chapter’s PhD-level structure is not the complexity of any single algorithm; it is the systematic integration and causal evaluation. The notebook runs an ablation suite: remove decay, remove novelty, remove contradiction, remove retrieval, remove belief. This is how you learn which component prevents which failure mode. If removing novelty increases turnover and drawdowns, you have evidence that redundancy control reduces churn-induced execution debt. If removing contradiction increases “confidently wrong” behavior—high retrieval confidence while being directionally wrong—you have quantified a failure mode that resembles confident hallucination. If removing belief collapses regime alignment despite retrieval, you have shown that memory is not a substitute for latent state inference. This is the kind of causal decomposition that makes the results teachable and reviewable.

Robustness is then evaluated through a stress surface. Volatility and jump intensity are increased systematically, and the pipeline is rerun end-to-end. This is a core professional behavior: you do not trust a system until you have mapped its failure surface. The notebook’s stress suite produces explicit failure flags (drawdown, equity collapse, churn), and those flags are written into a risk log as machine-readable artifacts. This matters for governance. A risk log is not a narrative. It is a structured object tied to evidence. It can be compared across runs, tracked over time, and reviewed formally.

Finally, the chapter is governance-first in a strict operational sense. It writes a run manifest, events log, diagnostics artifacts, stress results, risk log, verification register, and a consolidated governance store, then packages everything into a hash-verified audit bundle. This is not bureaucracy. It is the minimum infrastructure required to make temporal agents acceptable in finance. If an agent’s decisions cannot be reconstructed—what belief it held, what evidence it retrieved, what gates it applied, what constraints bound it—then the agent is a gadget. With traceability and controls, the agent becomes an instrument: powerful, auditable, and reviewable.

Chapter 3 therefore reframes temporal intelligence as the disciplined management of history. Memory without governance is accumulation. Belief without diagnostics is overconfidence. Retrieval without gating is interference. Policy without execution realism is fiction. The purpose of this chapter is to put all of those truths into one governed laboratory: a system where long context is not celebrated as a feature, but handled as a risk surface to be measured, controlled, and made auditable.

##2.SYNTHETIC MARFKET GENERATOR

###2.1.OVERVIEW

**Cell 2 — Synthetic market with latent regimes, heavy tails, jumps, and execution frictions**  
Cell 2 constructs the environment in which temporal intelligence will be tested: a synthetic market where the truth is known but the observations are realistic enough to generate ambiguity. The market is driven by a latent regime process modeled as a Markov chain (an HMM where the state is unobserved). Each regime has its own drift and volatility, but the return innovations are not Gaussian. Instead, returns are Student-t distributed to create heavy tails and occasional extreme moves, with degrees of freedom regime-dependent to encode “quiet” versus “stress” phases. A jump component adds rare but impactful discontinuities, with intensities and magnitudes varying by regime. This creates a structured mismatch between “normal times” and “crisis times,” which is precisely where temporal inference becomes meaningful and fragile.  

The cell also defines regime-dependent spreads (in basis points) and widens spreads when jumps occur. This may look like a small addition, but it is critical for the chapter’s finance rigor. Without execution frictions, a policy can trade aggressively without paying the economic price. Here, the environment explicitly links stress to execution cost: shocks widen spreads, which increases the cost of turnover. This is the foundation for later conclusions about execution debt and capacity cliffs.  

The outputs include a price path (via cumulative log returns), the latent regime path (ground truth), a jump indicator series, and the spread series. The latent regime path is essential: it allows you to measure how well belief inference works and whether the agent’s policy aligns with the true drift sign of the regime. The cell plots both price and regime to provide immediate sanity checking: you should see visually that regimes persist, jumps occur, and prices exhibit fat-tailed behavior.  

This cell introduces the chapter’s central epistemic structure: partial observability. The agent will never directly observe the regime state, but the environment is constructed so that regimes imprint themselves on the distribution of returns and execution conditions. This is exactly the scenario faced in real finance: “regime” is a latent construct, inferred from noisy signals under nonstationarity. By generating the market synthetically, the notebook avoids false claims while still producing a world rich enough to make advanced inference and governance meaningful. Cell 2 therefore defines the controlled battlefield: the agent must infer hidden structure, act under constraints, and remain auditable.


###2.2.CODE AND IMPLEMENTATION

```python
# =============================================================================
# CELL 2 — Synthetic Market Generator (Regime-HMM + Heavy Tails + Jumps + Spread) + Ground Truth
# =============================================================================
def assert_row_stochastic(P: np.ndarray, tol: float = 1e-10) -> None:
    rs = np.sum(P, axis=1)
    if np.max(np.abs(rs - 1.0)) > tol:
        raise ValueError(f"Transition matrix rows must sum to 1. Max deviation={float(np.max(np.abs(rs-1.0)))}")

def sample_markov_path(P: np.ndarray, T: int, s0: int, rng: np.random.Generator) -> np.ndarray:
    S = P.shape[0]
    s = np.zeros(T, dtype=int)
    s[0] = s0
    for t in range(1, T):
        s[t] = int(rng.choice(S, p=P[s[t-1]]))
    return s

def student_t_std(df: float) -> float:
    # variance of t(df) is df/(df-2) for df>2
    if df <= 2.0:
        raise ValueError("df must be > 2 for finite variance")
    return math.sqrt(df / (df - 2.0))

def generate_market(cfg: ExperimentConfig, rng: np.random.Generator,
                    vol_mult: float = 1.0, jump_mult: float = 1.0) -> Dict[str, np.ndarray]:
    P = np.array(cfg.P, dtype=float)
    assert_row_stochastic(P)

    s0 = int(rng.integers(0, cfg.S))
    regimes = sample_markov_path(P, cfg.T, s0, rng)

    r = np.zeros(cfg.T, dtype=float)
    jumps = np.zeros(cfg.T, dtype=int)
    spreads_bps = np.zeros(cfg.T, dtype=float)

    for t in range(cfg.T):
        st = int(regimes[t])
        mu = cfg.mu[st] * cfg.dt
        sig = cfg.sigma[st] * math.sqrt(cfg.dt) * vol_mult

        df = cfg.df_t[st]
        z = rng.standard_t(df) / (student_t_std(df) + 1e-12)  # variance-normalized
        base = mu + sig * float(z)

        lam = cfg.jump_lambda[st] * jump_mult
        if rng.random() < lam:
            jumps[t] = 1
            j = rng.normal(loc=-0.5 * cfg.jump_scale[st], scale=cfg.jump_scale[st] * jump_mult)
            base += float(j)

        r[t] = base
        spreads_bps[t] = cfg.spread_bps[st] * (1.0 + 2.5 * jumps[t])  # widen on jumps

    # price (log)
    logp = np.zeros(cfg.T + 1, dtype=float)
    logp[0] = 0.0
    logp[1:] = np.cumsum(r)

    p = np.exp(logp)
    return {
        "regime": regimes,
        "ret": r,
        "logp": logp,
        "price": p,
        "jumps": jumps,
        "spread_bps": spreads_bps,
    }

rng = np.random.default_rng(CFG.seed)
market = generate_market(CFG, rng, vol_mult=1.0, jump_mult=1.0)

log_event("market_generated", {"T": CFG.T, "S": CFG.S, "note": "HMM+Student-t+jumps created"})
print("Market shapes:", {k: v.shape for k, v in market.items()})

# Quick sanity visualization
plt.figure(figsize=(14,3.5))
plt.plot(market["price"], linewidth=1.2)
plt.title("Synthetic Price Path (Regime HMM + Heavy Tails + Jumps)")
plt.xlabel("t")
plt.ylabel("price")
plt.show()

plt.figure(figsize=(14,2.5))
plt.plot(market["regime"], drawstyle="steps-post")
plt.title("Latent Regime Path (Ground Truth)")
plt.xlabel("t")
plt.ylabel("regime")
plt.show()
```

##3.EPISODIC EVENT STREAM

###3.1.OVERVIEW

**Cell 3 — Episodic event stream: semantic memory under time, coupling, and adversarial contamination**  
Cell 3 creates the second information channel that distinguishes Chapter 3 from Chapter 2: an episodic memory stream that behaves like long-context text evidence but is represented purely as embeddings. Each event is a time-stamped vector in an embedding space with K topics. The topic distribution is regime-coupled: different regimes produce different mixtures of topics, which makes the event stream informative about the latent state. This is a deliberate design choice: it means “context” is not independent noise. It is a temporally structured evidence process.  

Each event also carries a polarity scalar, interpreted as directional or sentiment-like information. Polarity is constructed as a topic-level baseline with mild coupling to the realized return sign. This produces realistic ambiguity: events sometimes agree with market direction, sometimes lag it, and sometimes contradict it. That ambiguity is crucial because the agent will later need gating logic to avoid conditioning on misleading evidence.  

The cell then injects two forms of contamination. Near-duplicates replicate events with tiny perturbations. This models redundancy and collision in long context: multiple items that are essentially the same claim can dominate retrieval neighborhoods and create hubness. Hard negatives are the adversarial case: events are generated to be extremely aligned with a fixed query direction but are not genuinely relevant. This models the “false friend” problem in finance: documents that match keywords and embeddings but refer to the wrong entity, wrong time, wrong jurisdiction, or wrong regime. The injection rates are explicit parameters, which allows later stress testing and phase transition analysis.  

The event stream is sorted by time, and the notebook records event intensity across time. This matters because context is not static: evidence arrives in bursts, and bursts are exactly when agents are most likely to overreact. Cell 3 therefore sets up the temporal challenge: the agent will have a long archive of episodic events, but what matters is which events are relevant now, how quickly relevance decays, and how to avoid adversarial alignment. In short, Cell 3 converts “long context” into a governable memory object: time-stamped, semantically structured, partially informative, and explicitly contaminated in ways that resemble real-world retrieval risk.


###3.2.CODE AND IMPLEMENTATION

```python
# =============================================================================
# CELL 3 — Episodic Event Stream (Embedding Topics + Time Stamps + Hard Negatives + Near-Duplicates)
# =============================================================================
def l2_normalize_rows(X: np.ndarray, eps: float = 1e-12) -> np.ndarray:
    n = np.linalg.norm(X, axis=1, keepdims=True)
    return X / (n + eps)

def generate_topic_centroids(K: int, D: int, sep: float, rng: np.random.Generator) -> np.ndarray:
    C = rng.normal(size=(K, D)).astype(float)
    C = l2_normalize_rows(C)
    C *= sep
    C = l2_normalize_rows(C)
    return C

@dataclass
class Event:
    t: int
    embed: np.ndarray     # (D,)
    topic: int            # ground truth topic id
    polarity: float       # "sentiment" / directional hint (-1..+1)
    kind: str             # "normal" | "dup" | "hard_negative"

def build_event_stream(cfg: ExperimentConfig, market: Dict[str, np.ndarray], rng: np.random.Generator) -> Dict[str, Any]:
    """
    Generate episodic events whose topic distribution depends on latent regimes
    (temporal coupling) to create meaningful temporal intelligence tasks.
    """
    C = generate_topic_centroids(cfg.K_topics, cfg.D_embed, cfg.topic_sep, rng)

    # Regime→topic mixture map (hand-designed to create structural coupling)
    # Each regime activates a different subset of topics.
    S = cfg.S
    K = cfg.K_topics
    mix = np.zeros((S, K), dtype=float)
    for s in range(S):
        base = rng.random(K) * 0.2
        hot = [(3*s + i) % K for i in range(3)]
        base[hot] += 1.0
        mix[s] = base / np.sum(base)

    events: List[Event] = []
    topic_pol = rng.normal(loc=0.0, scale=0.8, size=K).astype(float)
    topic_pol = np.tanh(topic_pol)  # clamp to (-1,1)

    # query direction: pretend the agent cares about topics most aligned to "fundamental risk"
    # We create a query vector q_event used later in retrieval evaluation.
    q_event = rng.normal(size=cfg.D_embed).astype(float)
    q_event = q_event / (np.linalg.norm(q_event) + 1e-12)

    for t in range(cfg.T):
        # Poisson number of events per step
        n_events = rng.poisson(cfg.event_rate)
        if n_events <= 0:
            continue

        s = int(market["regime"][t])
        probs = mix[s]

        for _ in range(int(n_events)):
            topic = int(rng.choice(K, p=probs))
            e = C[topic] + rng.normal(scale=cfg.topic_sigma, size=cfg.D_embed).astype(float)
            e = e / (np.linalg.norm(e) + 1e-12)

            pol = float(topic_pol[topic] + 0.20 * np.sign(market["ret"][t]))
            pol = float(np.clip(pol, -1.0, 1.0))
            events.append(Event(t=t, embed=e, topic=topic, polarity=pol, kind="normal"))

    # Inject near-duplicates
    Mdup = int(cfg.event_dup_rate * max(1, len(events)))
    if Mdup > 0:
        idxs = rng.integers(0, len(events), size=Mdup)
        for i in idxs:
            base = events[int(i)]
            e = base.embed + rng.normal(scale=0.01, size=cfg.D_embed).astype(float)
            e = e / (np.linalg.norm(e) + 1e-12)
            events.append(Event(t=base.t, embed=e, topic=base.topic, polarity=base.polarity, kind="dup"))

    # Inject hard negatives: aligned to q_event but wrong topic
    Madv = int(cfg.event_adversary_rate * max(1, len(events)))
    if Madv > 0:
        for _ in range(Madv):
            t = int(rng.integers(0, cfg.T))
            # extremely close to q_event
            e = q_event + rng.normal(scale=0.004, size=cfg.D_embed).astype(float)
            e = e / (np.linalg.norm(e) + 1e-12)
            topic = int(rng.integers(0, cfg.K_topics))
            pol = float(np.clip(rng.normal(loc=0.0, scale=0.6), -1.0, 1.0))
            events.append(Event(t=t, embed=e, topic=topic, polarity=pol, kind="hard_negative"))

    # sort by time
    events.sort(key=lambda x: x.t)

    return {
        "topic_centroids": C,
        "regime_topic_mix": mix,
        "events": events,
        "q_event": q_event,
        "topic_polarity": topic_pol,
    }

event_world = build_event_stream(CFG, market, rng)
events: List[Event] = event_world["events"]
log_event("event_stream_built", {"event_count": len(events), "K_topics": CFG.K_topics, "D": CFG.D_embed})
print("Events:", len(events), "| first_t:", events[0].t if events else None)

# Visual: event count per time bucket
counts = np.zeros(CFG.T, dtype=int)
for e in events:
    counts[e.t] += 1

plt.figure(figsize=(14,3))
plt.plot(counts, linewidth=1.0)
plt.title("Episodic Event Intensity (per time step)")
plt.xlabel("t")
plt.ylabel("count")
plt.show()
```

##4.PARTICLE FILTER FOR LATENT REGIME

###4.1.OVERVIEW

**Cell 4 — Particle-filter belief over latent regimes plus CUSUM mismatch alarms**  
Cell 4 constructs the belief state that makes the agent truly temporal rather than reactive. The environment has a latent regime that controls drift, volatility, tails, and jump intensity. The agent does not observe that regime; it observes only returns. Therefore, the agent must maintain a posterior distribution over regimes, updated sequentially as new observations arrive. This cell implements a particle filter: a Monte Carlo approximation to Bayesian filtering. Particles represent candidate regime states; they propagate forward through the Markov transition matrix and are reweighted by the likelihood of the observed return under each regime’s Student-t observation model.  

The use of a Student-t likelihood is an advanced choice. It matches the heavy-tailed return generator and prevents Gaussian likelihoods from over-penalizing extreme returns, which would otherwise cause posterior collapse. The implementation emphasizes numerical stability: log-likelihoods are normalized with log-sum-exp, and resampling is triggered when effective sample size (ESS) falls below a threshold. ESS is a governance-critical diagnostic: it tells you whether your filter is degenerating and whether your posterior is dominated by a few particles.  

The cell also adds a second layer: CUSUM alarms on standardized residuals. Even if the filter is mathematically correct, models can be misspecified. CUSUM provides a fast, interpretable detector of sustained mismatch: if standardized residuals accumulate beyond a threshold, the cell emits an alarm. This is not merely a signal; it is a governance control. In practice, an alarm would trigger “do not trust the model” behaviors: reduce leverage, freeze retraining, require human review, or switch to a conservative policy. Here, it becomes part of the trace that later diagnostics can examine.  

The output includes posterior probabilities over regimes, ESS over time, and alarm events. Visualizations allow immediate inspection: posterior trajectories should track regime shifts with lag; alarms should cluster around structural breaks and jump clusters. This cell therefore provides the agent’s epistemic core: not a single regime guess, but a full uncertainty representation. Finance rigor requires uncertainty awareness: acting on point estimates under nonstationarity is a recipe for blowups. Cell 4 makes uncertainty explicit, measurable, and auditable, enabling the later agent to condition its behavior on belief and alarms rather than raw observations alone.


###4.2.CODE AND IMPLEMENTATION

```python
# =============================================================================
# CELL 4 (FIXED) — Particle Filter for Latent Regime (Belief State) + CUSUM Change-Point Signal
#   Fixes common CH3 Cell-4 breakpoints:
#   - removes walrus operator (compatibility)
#   - hardens numerical stability (log-sum-exp)
#   - validates transition matrix shape + stochastic rows
#   - avoids silent NaNs / inf in likelihood scaling
# =============================================================================
import math
from dataclasses import dataclass

def assert_row_stochastic(P: np.ndarray, tol: float = 1e-10) -> None:
    P = np.asarray(P, dtype=float)
    if P.ndim != 2 or P.shape[0] != P.shape[1]:
        raise ValueError(f"P must be square. Got shape={P.shape}")
    rs = np.sum(P, axis=1)
    if np.max(np.abs(rs - 1.0)) > tol:
        raise ValueError(f"Transition matrix rows must sum to 1. Max deviation={float(np.max(np.abs(rs-1.0)))}")
    if np.any(P < -tol):
        raise ValueError("Transition matrix contains negative probabilities.")

def log_student_t_pdf(x: float, df: float, loc: float, scale: float) -> float:
    # Stable Student-t logpdf for df>2, scale>0
    if df <= 2.0:
        raise ValueError(f"df must be > 2 for finite variance. df={df}")
    if scale <= 0.0:
        raise ValueError(f"scale must be > 0. scale={scale}")

    z = (x - loc) / scale
    a = math.lgamma((df + 1.0) / 2.0) - math.lgamma(df / 2.0)
    b = -0.5 * math.log(df * math.pi) - math.log(scale)
    c = -((df + 1.0) / 2.0) * math.log1p((z * z) / df)   # log1p for stability
    return a + b + c

def systematic_resample(weights: np.ndarray, rng: np.random.Generator) -> np.ndarray:
    w = np.asarray(weights, dtype=float)
    N = w.size
    if N == 0:
        return np.array([], dtype=int)
    positions = (rng.random() + np.arange(N)) / N
    cumsum = np.cumsum(w)
    idx = np.zeros(N, dtype=int)
    i = 0
    j = 0
    while i < N:
        if positions[i] < cumsum[j]:
            idx[i] = j
            i += 1
        else:
            j += 1
            if j >= N:
                # numerical guard: in case cumsum[-1] < 1 due to rounding
                idx[i:] = N - 1
                break
    return idx

def logsumexp(a: np.ndarray) -> float:
    a = np.asarray(a, dtype=float)
    m = float(np.max(a))
    return m + float(np.log(np.sum(np.exp(a - m)) + 1e-300))

@dataclass
class BeliefTrace:
    pi: np.ndarray         # (T,S)
    ess: np.ndarray        # (T,)
    cusum_pos: np.ndarray  # (T,)
    cusum_neg: np.ndarray  # (T,)
    alarm: np.ndarray      # (T,)

def particle_filter_regime(cfg: ExperimentConfig, market: Dict[str, np.ndarray], rng: np.random.Generator) -> BeliefTrace:
    P = np.array(cfg.P, dtype=float)
    assert_row_stochastic(P)
    if P.shape[0] != cfg.S:
        raise ValueError(f"cfg.S={cfg.S} but P is {P.shape[0]}x{P.shape[1]}")

    T = int(cfg.T)
    S = int(cfg.S)
    Np = int(cfg.particle_N)

    # initialize particles uniformly over regimes
    particles = rng.integers(0, S, size=Np).astype(int)
    w = np.ones(Np, dtype=float) / Np

    pi = np.zeros((T, S), dtype=float)
    ess = np.zeros(T, dtype=float)

    cusum_pos = np.zeros(T, dtype=float)
    cusum_neg = np.zeros(T, dtype=float)
    alarm = np.zeros(T, dtype=int)

    gpos = 0.0
    gneg = 0.0

    # Precompute per-regime constants for speed and stability
    mu = np.array(cfg.mu, dtype=float) * float(cfg.dt)
    sig = np.array(cfg.sigma, dtype=float) * math.sqrt(float(cfg.dt))
    df = np.array(cfg.df_t, dtype=float)

    for t in range(T):
        # propagate
        # (vectorized propagation is tricky due to state-dependent probs; do fast loop)
        for i in range(Np):
            s_prev = int(particles[i])
            particles[i] = int(rng.choice(S, p=P[s_prev]))

        # weight
        r_obs = float(market["ret"][t])
        logw = np.empty(Np, dtype=float)
        for i in range(Np):
            s = int(particles[i])
            scale = float(sig[s] + 1e-12)
            logw[i] = log_student_t_pdf(r_obs, float(df[s]), float(mu[s]), scale)

        # normalize via log-sum-exp
        lse = logsumexp(logw)
        w = np.exp(logw - lse)
        w_sum = float(np.sum(w))
        if not np.isfinite(w_sum) or w_sum <= 0.0:
            # hard fallback to uniform (governed fail-safe)
            w = np.ones(Np, dtype=float) / Np
            log_event("belief_weight_fallback_uniform", {"t": int(t), "reason": "nonfinite_weights"})
        else:
            w = w / (w_sum + 1e-12)

        # posterior over regimes
        for s in range(S):
            pi[t, s] = float(np.sum(w[particles == s]))

        # ESS
        ess_t = 1.0 / (float(np.sum(w * w)) + 1e-12)
        ess[t] = ess_t

        # resample when needed
        if ess_t < float(cfg.particle_resample_thresh) * Np:
            idx = systematic_resample(w, rng)
            particles = particles[idx]
            w = np.ones(Np, dtype=float) / Np

        # CUSUM on standardized residual using MAP regime
        s_hat = int(np.argmax(pi[t]))
        mu_hat = float(mu[s_hat])
        sig_hat = float(sig[s_hat] + 1e-12)
        z = (r_obs - mu_hat) / sig_hat

        gpos = max(0.0, gpos + z - float(cfg.cusum_k))
        gneg = min(0.0, gneg + z + float(cfg.cusum_k))

        cusum_pos[t] = gpos
        cusum_neg[t] = gneg

        if (gpos > float(cfg.cusum_h)) or (abs(gneg) > float(cfg.cusum_h)):
            alarm[t] = 1
            gpos = 0.0
            gneg = 0.0

    return BeliefTrace(pi=pi, ess=ess, cusum_pos=cusum_pos, cusum_neg=cusum_neg, alarm=alarm)

# Run belief filter
belief = particle_filter_regime(CFG, market, rng)

(AUDIT_DIR / "belief_trace_summary.json").write_text(
    json.dumps({
        "run_id": RUN_ID,
        "verification_status": "Not Verified",
        "pi_last": belief.pi[-1].tolist(),
        "ess_min": float(np.min(belief.ess)),
        "alarms_count": int(np.sum(belief.alarm)),
    }, indent=2),
    encoding="utf-8"
)
log_event("belief_filter_done", {"alarms": int(np.sum(belief.alarm)), "ess_min": float(np.min(belief.ess))})

# Visualize belief vs truth
plt.figure(figsize=(14,3))
for s in range(CFG.S):
    plt.plot(belief.pi[:, s], label=f"P(s={s})")
plt.title("Particle Filter Belief Over Regimes")
plt.xlabel("t")
plt.ylabel("probability")
plt.legend()
plt.show()

plt.figure(figsize=(14,2.5))
plt.plot(belief.alarm, drawstyle="steps-post")
plt.title("CUSUM Alarms (Model Mismatch / Change-Point Signal)")
plt.xlabel("t")
plt.ylabel("alarm")
plt.show()
```

##5.TEMPORAL EPISODIC MEMORY STORE

###5.1.OVERVIEW

**Cell 5 — Governed episodic memory: decay, novelty gating, contradiction penalties, retrieval budgets**  
Cell 5 builds the memory system that transforms raw episodic events into controlled evidence selection. The memory store contains time-stamped embedding vectors with polarity and provenance tags (“normal,” “dup,” “hard_negative”). Retrieval is not a naive top-k cosine search. Instead, the score is a composite: cosine relevance is multiplied by temporal decay and by gating penalties that encode governance. Temporal decay implements the basic epistemic principle that older evidence should matter less, unless explicitly preserved by design. The decay function is parameterized by a half-life, making it interpretable and tunable.  

Novelty gating addresses hubness and duplication. Long context often contains repeated claims, recycled templates, and redundant summaries. If retrieval is purely similarity-based, redundancy can dominate the top of the list, creating an illusion of “strong evidence” while actually amplifying one source. Novelty gating penalizes high-density items using a local similarity density proxy: items that are near-duplicates of many others receive lower weights. This is a principled response to the “copy-paste consensus” failure mode common in finance memos and market commentary.  

Contradiction gating penalizes evidence that conflicts with the agent’s belief-driven expected drift. This is not censorship; it is controlled skepticism. If the belief state suggests a stress regime with negative drift, strongly positive polarity evidence should be down-weighted unless it is unusually relevant and recent. Contradiction gating therefore prevents the memory from whipsawing the policy under noisy sentiment bursts. It is explicitly parameterized, allowing the ablation study in Cell 6 to quantify its contribution.  

A key professional constraint introduced here is the retrieval budget, a proxy for context window limits. Even if memory holds thousands of events, the agent can condition only on a small bounded set. The notebook models this explicitly by truncating retrieved items to a fixed budget and renormalizing weights. This makes tradeoffs real: more retrieval increases coverage but raises interference risk; less retrieval increases stability but may miss crucial evidence.  

Cell 5 therefore instantiates a governed memory architecture: relevance is necessary but insufficient; time, redundancy, and contradiction must be controlled. The cell produces retrieval outputs with weights and notes that quantify density, expected drift, and budget usage. This bridges Chapter 2’s geometry with Chapter 3’s temporal reasoning: memory is not merely stored; it is selectively surfaced under governance.


###5.2.CODE AND IMPLEMENTATION

```python
# =============================================================================
# CELL 5 — Temporal Episodic Memory Store (Decay, Novelty, Contradiction) + Retrieval/Gating
# =============================================================================
def half_life_decay(age: float, half_life: float) -> float:
    # decay(age) = 2^(-age/half_life)
    return 2.0 ** (-age / (half_life + 1e-12))

def softmax(x: np.ndarray, temp: float) -> np.ndarray:
    z = x * float(temp)
    z = z - np.max(z)
    e = np.exp(z)
    return e / (np.sum(e) + 1e-12)

@dataclass
class MemoryItem:
    t: int
    embed: np.ndarray    # (D,)
    polarity: float
    kind: str
    topic: int

@dataclass
class RetrievalResult:
    idx: np.ndarray
    weights: np.ndarray
    score: np.ndarray
    budget_used: int
    notes: Dict[str, Any]

class EpisodicMemory:
    """
    A governed episodic memory:
      - stores events with time stamps
      - retrieval uses relevance (cosine) * temporal decay * gating penalties
      - returns weighted set under a fixed budget (proxy for context window)
    """
    def __init__(self, D: int, half_life: float, novelty_penalty: float, contradiction_penalty: float):
        self.D = D
        self.half_life = half_life
        self.novelty_penalty = novelty_penalty
        self.contradiction_penalty = contradiction_penalty
        self.items: List[MemoryItem] = []

    def add(self, e: Event) -> None:
        self.items.append(MemoryItem(t=e.t, embed=e.embed.astype(float), polarity=float(e.polarity), kind=e.kind, topic=int(e.topic)))

    def _as_matrix(self) -> Tuple[np.ndarray, np.ndarray, np.ndarray, List[str]]:
        if not self.items:
            return np.zeros((0, self.D)), np.zeros(0), np.zeros(0), []
        X = np.stack([it.embed for it in self.items], axis=0)
        ts = np.array([it.t for it in self.items], dtype=float)
        pol = np.array([it.polarity for it in self.items], dtype=float)
        kinds = [it.kind for it in self.items]
        return X, ts, pol, kinds

    def retrieve(
        self,
        t_now: int,
        q: np.ndarray,
        k: int,
        budget: int,
        gate_temp: float,
        belief_regime: np.ndarray,
        topic_centroids: np.ndarray,
        topic_polarity: np.ndarray
    ) -> RetrievalResult:
        """
        Returns weighted top-k items under a retrieval budget.
        Gating includes:
          - temporal decay
          - novelty penalty (prefers diversity over duplicates/hubs)
          - contradiction penalty (items whose implied polarity contradicts belief-weighted expected drift)
        """
        q = np.asarray(q, dtype=float).reshape(-1)
        q = q / (np.linalg.norm(q) + 1e-12)

        X, ts, pol, kinds = self._as_matrix()
        if X.shape[0] == 0:
            return RetrievalResult(idx=np.array([], dtype=int), weights=np.array([]), score=np.array([]),
                                   budget_used=0, notes={"empty": True})

        # base relevance
        sim = X @ q

        # temporal decay
        ages = float(t_now) - ts
        decay = np.array([half_life_decay(age=a, half_life=self.half_life) for a in ages], dtype=float)

        # novelty penalty: penalize near-duplicates by local density around each item (approx hubness)
        # We do a cheap proxy: similarity to a small random subset of memory items
        n = X.shape[0]
        m = min(180, n)
        idx_sub = np.random.choice(n, size=m, replace=False) if n > 0 else np.array([], dtype=int)
        if m >= 2:
            S = X @ X[idx_sub].T
            # density proxy = fraction of sims above threshold
            dens = np.mean(S > 0.92, axis=1)
        else:
            dens = np.zeros(n, dtype=float)
        novelty_gate = np.exp(-self.novelty_penalty * dens)

        # contradiction penalty: compare item polarity vs belief-weighted expected return sign
        # Build expected sign from belief over regimes using mu.
        # Here we do a proxy using the belief itself (passed in) and a fixed mapping:
        expected = float(np.dot(belief_regime, np.array(CFG.mu, dtype=float)))
        expected_sign = 0.0 if abs(expected) < 1e-8 else float(np.sign(expected))
        contradiction = (expected_sign * pol) < 0.0
        contra_gate = np.exp(-self.contradiction_penalty * contradiction.astype(float))

        # kind penalty (hard negatives and dups are more suspicious)
        kind_pen = np.ones(n, dtype=float)
        for i, knd in enumerate(kinds):
            if knd == "hard_negative":
                kind_pen[i] *= 0.35
            elif knd == "dup":
                kind_pen[i] *= 0.70

        score = sim * decay * novelty_gate * contra_gate * kind_pen

        # select top-k by score
        kk = min(k, n)
        idx = np.argpartition(score, -kk)[-kk:]
        idx = idx[np.argsort(score[idx])[::-1]]

        # weights for conditioning (softmax)
        w = softmax(score[idx], temp=gate_temp)

        # budget usage (proxy): assume each item costs 1 token; cap by budget
        # If k > budget, truncate.
        if idx.size > budget:
            idx = idx[:budget]
            w = w[:budget]
            w = w / (np.sum(w) + 1e-12)

        notes = {
            "n_items": int(n),
            "mean_decay": float(np.mean(decay)),
            "mean_density": float(np.mean(dens)),
            "expected_mu": expected,
            "expected_sign": expected_sign,
            "budget": int(budget),
        }
        return RetrievalResult(idx=idx.astype(int), weights=w.astype(float), score=score[idx].astype(float),
                               budget_used=int(idx.size), notes=notes)

# build memory and ingest all events once (agent will "see" events as time advances, but store exists)
mem = EpisodicMemory(D=CFG.D_embed,
                     half_life=CFG.event_decay_half_life,
                     novelty_penalty=CFG.novelty_penalty,
                     contradiction_penalty=CFG.contradiction_penalty)

for e in events:
    mem.add(e)

log_event("memory_initialized", {"stored_items": len(mem.items), "half_life": CFG.event_decay_half_life})
print("Memory items stored:", len(mem.items))

# quick retrieval demo at mid time
t_demo = CFG.T // 2
q_event = event_world["q_event"]
belief_demo = belief.pi[t_demo]
rr = mem.retrieve(
    t_now=t_demo,
    q=q_event,
    k=CFG.retrieval_topk,
    budget=CFG.retrieval_budget,
    gate_temp=CFG.gate_temperature,
    belief_regime=belief_demo,
    topic_centroids=event_world["topic_centroids"],
    topic_polarity=event_world["topic_polarity"],
)
print("Retrieval demo:", rr.notes, "| budget_used:", rr.budget_used)
```

##6.GOVERNED TEMPORAL AGENT

###6.1.OVERVIEW

**Cell 6 — Temporal agent: belief + memory retrieval + policy + execution realism + ablation suite**  
Cell 6 is the chapter’s central systems integration: a temporal agent that combines latent belief inference with governed episodic retrieval and executes a constrained trading policy with realistic frictions. The agent’s state at time t includes (i) the posterior belief over regimes, (ii) the retrieved episodic evidence set and its weights, and (iii) operational risk signals such as CUSUM alarms. The policy computes a raw signal as a nonlinear blend of belief-driven expected drift and memory-derived evidence polarity. The nonlinearities (tanh scaling) are deliberate: they prevent extreme raw signals from dominating and reflect the idea that marginal evidence has diminishing influence.  

Finance rigor enters through risk controls and execution. The agent applies volatility targeting using an EWMA volatility estimator: position size is scaled inversely with estimated volatility to maintain a target risk budget. Leverage is capped to enforce feasibility, and turnover is capped to represent operational limits and to avoid pathological churn. Costs are modeled as the sum of spread, fees, and convex impact. Convex impact is critical: it creates capacity cliffs where aggressive turnover becomes disproportionately expensive, mirroring real market microstructure.  

The ablation suite is the PhD-level experimental design. Instead of presenting one agent, the notebook defines multiple agents that remove one control at a time: no decay, no novelty gating, no contradiction gating, no retrieval, and no belief. This allows causal interpretation. If removing novelty gating increases churn and drawdowns, you have evidence that redundancy control matters. If removing contradiction gating increases “confidently wrong” behavior, you can quantify how evidence conflict destabilizes policy. If removing belief collapses performance despite retrieval, you learn that memory alone is not a substitute for a structured latent state.  

Cell 6 logs each agent run and records final equity outcomes. The outputs include full time series: equity, position, turnover, costs, PnL increments, evidence mass, gate confidence proxy, inferred regime, and alarms. This is not just backtesting; it is a controlled experiment in temporal intelligence under governance. The result is a complete trace of how belief, memory, and controls shape behavior. This trace becomes the foundation for the diagnostics in Cell 7 and the stress surfaces in Cell 8, enabling professional-grade interpretation rather than anecdotal conclusions.


###6.2.CODE AND IMPLEMENTATION

```python
# =============================================================================
# CELL 6 — Governed Temporal Agent: Belief + Retrieval + Policy + Execution + Risk Controls
# =============================================================================
@dataclass
class AgentConfig:
    name: str
    use_belief: bool
    use_temporal_decay: bool
    use_novelty_gate: bool
    use_contradiction_gate: bool
    use_retrieval: bool

@dataclass
class AgentTrace:
    equity: np.ndarray        # (T+1,)
    pos: np.ndarray           # (T,)
    turnover: np.ndarray      # (T,)
    cost: np.ndarray          # (T,)
    pnl_inc: np.ndarray       # (T,)
    retrieved_mass: np.ndarray# (T,) total weight on "normal" items (proxy)
    gate_conf: np.ndarray     # (T,) top score gap proxy
    s_hat: np.ndarray         # (T,) inferred regime
    alarms: np.ndarray        # (T,) cusum alarms

def ewma_vol(x: np.ndarray, lam: float = 0.97) -> np.ndarray:
    v = np.zeros_like(x)
    m2 = 0.0
    for i in range(x.size):
        m2 = lam * m2 + (1 - lam) * (x[i] ** 2)
        v[i] = math.sqrt(max(1e-12, m2))
    return v

def exec_cost(turnover: float, price: float, spread_bps: float, fee_bps: float, eta: float, gamma: float) -> float:
    spread = (spread_bps * 1e-4) * abs(turnover) * price
    fee = (fee_bps * 1e-4) * abs(turnover) * price
    impact = eta * (abs(turnover) ** gamma) * price
    return float(spread + fee + impact)

def clamp(x: float, lo: float, hi: float) -> float:
    return float(min(hi, max(lo, x)))

def run_agent(cfg: ExperimentConfig, agent_cfg: AgentConfig,
              market: Dict[str, np.ndarray], belief: BeliefTrace,
              mem: EpisodicMemory, event_world: Dict[str, Any],
              rng: np.random.Generator) -> AgentTrace:
    T = cfg.T
    eq = np.zeros(T + 1, dtype=float)
    pos = np.zeros(T, dtype=float)
    turnover = np.zeros(T, dtype=float)
    cost = np.zeros(T, dtype=float)
    pnl_inc = np.zeros(T, dtype=float)
    retrieved_mass = np.zeros(T, dtype=float)
    gate_conf = np.zeros(T, dtype=float)
    s_hat = np.zeros(T, dtype=int)
    alarms = belief.alarm.copy()

    eq[0] = 1.0
    p = market["price"]
    r = market["ret"]
    spread = market["spread_bps"]
    q_event = event_world["q_event"]

    # vol targeting
    vol_est = ewma_vol(r, lam=0.985)

    prev_pos = 0.0

    for t in range(T):
        # belief state
        if agent_cfg.use_belief:
            pi_t = belief.pi[t]
            s = int(np.argmax(pi_t))
        else:
            pi_t = np.ones(cfg.S, dtype=float) / cfg.S
            s = 0
        s_hat[t] = s

        # retrieval (episodic memory)
        if agent_cfg.use_retrieval:
            # toggle gating components by temporarily adjusting penalties/half-life
            # (governed ablation without editing memory store)
            original_hl = mem.half_life
            original_np = mem.novelty_penalty
            original_cp = mem.contradiction_penalty

            mem.half_life = cfg.event_decay_half_life if agent_cfg.use_temporal_decay else 1e12
            mem.novelty_penalty = cfg.novelty_penalty if agent_cfg.use_novelty_gate else 0.0
            mem.contradiction_penalty = cfg.contradiction_penalty if agent_cfg.use_contradiction_gate else 0.0

            rr = mem.retrieve(
                t_now=t,
                q=q_event,
                k=cfg.retrieval_topk,
                budget=cfg.retrieval_budget,
                gate_temp=cfg.gate_temperature,
                belief_regime=pi_t,
                topic_centroids=event_world["topic_centroids"],
                topic_polarity=event_world["topic_polarity"]
            )

            # restore
            mem.half_life = original_hl
            mem.novelty_penalty = original_np
            mem.contradiction_penalty = original_cp

            # compute a directional “evidence score” from retrieved polarity
            if rr.idx.size > 0:
                pols = np.array([mem.items[i].polarity for i in rr.idx], dtype=float)
                kinds = [mem.items[i].kind for i in rr.idx]
                # mass on normal events (proxy: how much clean evidence we used)
                retrieved_mass[t] = float(np.sum(rr.weights[np.array([k == "normal" for k in kinds])]))
                # confidence proxy: top score gap
                gate_conf[t] = float(rr.score[0] - rr.score[1]) if rr.score.size >= 2 else float(rr.score[0])
                evidence = float(np.dot(rr.weights, pols))
            else:
                evidence = 0.0
                retrieved_mass[t] = 0.0
                gate_conf[t] = 0.0
        else:
            evidence = 0.0

        # policy: combine belief-driven expected drift and retrieved evidence
        expected_mu = float(np.dot(pi_t, np.array(cfg.mu, dtype=float))) if agent_cfg.use_belief else 0.0
        raw_signal = 0.65 * math.tanh(18.0 * expected_mu) + 0.35 * math.tanh(2.2 * evidence)

        # risk control: volatility targeting
        vol = float(max(1e-8, vol_est[t]))
        scale = cfg.target_vol / vol
        desired = raw_signal * scale

        # leverage cap
        desired = clamp(desired, -cfg.leverage_cap, cfg.leverage_cap)

        # turnover cap
        delta = desired - prev_pos
        max_d = cfg.turnover_cap
        delta = clamp(delta, -max_d, max_d)
        new_pos = prev_pos + delta

        pos[t] = new_pos
        turnover[t] = abs(new_pos - prev_pos)

        # PnL: position earns return; costs charged on turnover
        pnl = prev_pos * r[t]
        c = exec_cost(turnover=turnover[t], price=p[t], spread_bps=spread[t],
                      fee_bps=cfg.fee_bps, eta=cfg.impact_eta, gamma=cfg.impact_gamma)
        cost[t] = c
        pnl_inc[t] = pnl - c / (p[t] + 1e-12)

        eq[t+1] = eq[t] * (1.0 + pnl_inc[t])

        prev_pos = new_pos

    return AgentTrace(eq, pos, turnover, cost, pnl_inc, retrieved_mass, gate_conf, s_hat, alarms)

# Define ablation suite (PhD-style: isolate contributions)
agents = [
    AgentConfig("A_full_governed", True, True, True, True, True),
    AgentConfig("B_no_decay", True, False, True, True, True),
    AgentConfig("C_no_novelty", True, True, False, True, True),
    AgentConfig("D_no_contra", True, True, True, False, True),
    AgentConfig("E_no_retrieval", True, True, True, True, False),
    AgentConfig("F_no_belief", False, True, True, True, True),
]

traces: Dict[str, AgentTrace] = {}
for a in agents:
    traces[a.name] = run_agent(CFG, a, market, belief, mem, event_world, rng)
    log_event("agent_run", {"agent": a.name, "final_equity": float(traces[a.name].equity[-1])})

# quick compare
for name, tr in traces.items():
    print(name, "final_eq=", float(tr.equity[-1]), "mean_turnover=", float(np.mean(tr.turnover)))
```

##7.DIAGNOSTICS

###7.1.OVERVIEW

**Cell 7 — Diagnostics: drawdowns, tail risk, regime alignment, evidence quality, confidence pathologies**  
Cell 7 converts raw agent traces into decision-relevant diagnostics. In finance, performance is not a single number; it is a surface of risks and failure modes. The cell computes maximum drawdown to measure survival risk: how deep and persistent the worst equity decline is. It computes Expected Shortfall at 95% on PnL increments to quantify tail risk beyond VaR-style quantiles, which is essential under heavy tails and jumps. It computes a Sharpe-like ratio for a rough risk-adjusted view, while acknowledging synthetic limitations.  

But the most important diagnostics are causal and governance-oriented. Regime alignment measures whether the agent’s position sign agrees with the true regime drift sign. This addresses the epistemic question: did the agent act in the right direction given latent conditions, or did it merely get lucky? Evidence mass measures how much weight the retrieval system placed on “normal” events versus contaminated events; it is a proxy for evidence quality under adversarial pressure. If evidence mass collapses, the system is likely being dominated by duplicates or hard negatives.  

The confidence pathology metric is the retrieval-to-policy analog of “confident hallucination.” It estimates the frequency with which the agent has high gating confidence (large top score gaps) while being directionally wrong relative to regime drift. This is a crucial professional control: a system that is wrong but uncertain may be acceptable under supervision; a system that is wrong and confident is dangerous. This metric operationalizes that distinction.  

The cell writes diagnostics to an artifact, enabling review and comparison across runs. It then plots equity curves for all ablation agents in one figure, allowing visual identification of instability modes: excessive churn, drawdown cliffs, or overfitting-like behavior. It also plots gate confidence for the governed agent to show when the memory system becomes decisive. The combination of numerical and visual diagnostics supports a model-risk style narrative: which control reduces which failure mode, and where confidence becomes misplaced. In short, Cell 7 is where the notebook stops being a simulation and becomes an evaluative instrument, translating temporal intelligence mechanics into risk metrics and governance signals.


###7.2.CODE AND IMPLEMENTATION

```python
# =============================================================================
# CELL 7 — Diagnostics: Tail Risk (ES), Drawdown, Evidence Quality, Regime Alignment, Confidence Pathologies
# =============================================================================
def max_drawdown(equity: np.ndarray) -> float:
    peak = equity[0]
    mdd = 0.0
    for x in equity:
        peak = max(peak, x)
        dd = (peak - x) / (peak + 1e-12)
        mdd = max(mdd, dd)
    return float(mdd)

def expected_shortfall(x: np.ndarray, alpha: float = 0.95) -> float:
    # x = pnl increments; ES on losses
    losses = -x
    q = float(np.quantile(losses, alpha))
    tail = losses[losses >= q]
    if tail.size == 0:
        return 0.0
    return float(np.mean(tail))

def sharpe_like(x: np.ndarray) -> float:
    m = float(np.mean(x))
    s = float(np.std(x)) + 1e-12
    return float(m / s)

def regime_alignment_score(pos: np.ndarray, mu: List[float], regime: np.ndarray) -> float:
    # sign alignment between pos and mu_s
    mu_arr = np.array(mu, dtype=float)
    mu_sign = np.sign(mu_arr[regime])
    pos_sign = np.sign(pos)
    # treat 0 as neutral, count only nonzero
    mask = (mu_sign != 0) & (pos_sign != 0)
    if np.sum(mask) == 0:
        return 0.0
    return float(np.mean((mu_sign[mask] == pos_sign[mask]).astype(float)))

def confidence_wrong_rate(tr: AgentTrace, market_regime: np.ndarray, mu: List[float], thresh: float = 0.015) -> float:
    # "confidently wrong": high gate_conf but position sign opposes regime drift sign
    mu_arr = np.array(mu, dtype=float)
    mu_sign = np.sign(mu_arr[market_regime])
    wrong = (np.sign(tr.pos) * mu_sign) < 0
    conf = tr.gate_conf > thresh
    mask = conf & wrong
    return float(np.mean(mask.astype(float)))

diag = {}
for name, tr in traces.items():
    diag[name] = {
        "final_equity": float(tr.equity[-1]),
        "mdd": max_drawdown(tr.equity),
        "es_95": expected_shortfall(tr.pnl_inc, 0.95),
        "sharpe_like": sharpe_like(tr.pnl_inc),
        "mean_turnover": float(np.mean(tr.turnover)),
        "mean_cost": float(np.mean(tr.cost)),
        "evidence_mass_mean": float(np.mean(tr.retrieved_mass)),
        "regime_alignment": regime_alignment_score(tr.pos, CFG.mu, market["regime"]),
        "confidently_wrong_rate": confidence_wrong_rate(tr, market["regime"], CFG.mu, thresh=0.010),
        "cusum_alarms": int(np.sum(tr.alarms)),
    }

(AUDIT_DIR / "diagnostics.json").write_text(json.dumps({"run_id": RUN_ID, "diagnostics": diag, "verification_status": "Not Verified"}, indent=2), encoding="utf-8")
log_event("diagnostics_written", {"file": "diagnostics.json"})

print(json.dumps(diag, indent=2))

# Visualization: equity curves
plt.figure(figsize=(14,4))
for name, tr in traces.items():
    plt.plot(tr.equity, label=name, linewidth=1.1)
plt.title("Equity Curves — Ablation Study (Temporal Intelligence Controls)")
plt.xlabel("t")
plt.ylabel("equity")
plt.legend(ncol=2)
plt.show()

# Confidence pathology visualization for one agent
ref = traces["A_full_governed"]
plt.figure(figsize=(14,3))
plt.plot(ref.gate_conf, linewidth=1.0)
plt.title("Gate Confidence Proxy (Top Score Gap) — A_full_governed")
plt.xlabel("t")
plt.ylabel("gap")
plt.show()
```

##8.STRESS TESTS

###8.1.OVERVIEW

**Cell 8 — Stress suite: volatility/jump surfaces, structural failure flags, and risk log integration)**  
Cell 8 tests robustness, not performance. It reruns the entire environment-agent pipeline under a grid of stress multipliers: volatility and jump intensity are increased systematically to create harsher regimes with wider spreads, heavier effective tails, and more discontinuities. Crucially, stress is not applied only to returns; it propagates through execution costs and belief filtering, which is realistic. In stressed environments, volatility targeting shrinks positions, but turnover constraints and convex impact can still create execution debt. Meanwhile, heavy tails can degrade belief inference and trigger CUSUM alarms.  

The stress suite is implemented as a surface: for each (vol_mult, jump_mult) pair, the notebook regenerates a new synthetic market, regenerates a new episodic event stream, refits belief via the particle filter, rebuilds memory, and runs every agent variant. This creates a structured dataset of outcomes rather than a single stress anecdote. The cell then defines explicit failure flags: drawdown beyond a threshold, final equity below a threshold, and churn above a threshold. These are not “performance judgments”; they are structural feasibility controls. In professional terms, they are stop-loss and model-risk triggers: the system is classified as failing under certain environmental stress states.  

The cell records each run into `stress_results.json` and writes a bounded list of flagged scenarios into `risk_log.json`. The risk log is not hand-written; it is generated as an artifact of stress testing. This is an important methodological shift: governance is evidence-backed. The risk log also lists risk taxonomies (belief misspecification, context interference, hard-negative dominance, execution debt) and references the controls introduced earlier (alarms, gating, caps, stress suites).  

Finally, the cell visualizes a stress surface for the governed agent’s final equity. This surface is not meant to “show green zones” for investing; it is meant to reveal where the agent’s design fails. The surface view is pedagogically powerful: it connects abstract parameters like jump intensity to concrete outcomes like equity collapse and churn spikes. In sum, Cell 8 is the robustness laboratory: it converts Chapter 3 into a professional-grade stress testing exercise where temporal intelligence is judged by stability under regime stress, not by an isolated backtest.


###8.2.CODE AND IMPLEMENTATION

```python
# =============================================================================
# CELL 8 (REPLACEMENT) — Stress Suite WITHOUT "Infinite Loop" Feel
#   - Scenario precompute (market/events/belief/memory) ONCE per stress point
#   - Progress + ETA
#   - Wall-clock + run-count guards
#   - Early stop on catastrophic failure
#   - Incremental artifact writing
# =============================================================================
from dataclasses import dataclass, asdict

@dataclass
class StressRow:
    vol_mult: float
    jump_mult: float
    agent: str
    final_equity: float
    mdd: float
    es_95: float
    mean_turnover: float
    fail_drawdown: int
    fail_equity: int
    fail_churn: int
    wall_sec: float

def _now() -> float:
    return time.time()

def _fmt_eta(sec: float) -> str:
    sec = max(0.0, float(sec))
    if sec < 60:
        return f"{sec:.0f}s"
    if sec < 3600:
        return f"{sec/60:.1f}m"
    return f"{sec/3600:.2f}h"

def compute_failure_flags(final_equity: float, mdd: float, mean_turnover: float) -> Tuple[int,int,int]:
    fail_dd = 1 if mdd > 0.35 else 0
    fail_eq = 1 if final_equity < 0.80 else 0
    fail_ch = 1 if mean_turnover > 0.22 else 0
    return fail_dd, fail_eq, fail_ch

# -----------------------------
# Hard guards (governed runtime)
# -----------------------------
MAX_STRESS_RUNS = 10_000           # absolute cap
MAX_WALL_SEC    = 180.0            # cap for this cell (set higher if you want)
EARLY_STOP_EQ   = 0.65             # catastrophic equity threshold (per scenario-agent)
EARLY_STOP_DD   = 0.55             # catastrophic drawdown threshold (per scenario-agent)

# -----------------------------
# Scenario cache: build ONCE per stress point
# -----------------------------
@dataclass
class StressScenario:
    vol_mult: float
    jump_mult: float
    market: Dict[str, np.ndarray]
    belief: BeliefTrace
    event_world: Dict[str, Any]
    memory: EpisodicMemory

def build_scenario(cfg: ExperimentConfig, rng: np.random.Generator, vol_mult: float, jump_mult: float) -> StressScenario:
    mk = generate_market(cfg, rng, vol_mult=vol_mult, jump_mult=jump_mult)
    ev = build_event_stream(cfg, mk, rng)
    bl = particle_filter_regime(cfg, mk, rng)

    mem_store = EpisodicMemory(cfg.D_embed, cfg.event_decay_half_life, cfg.novelty_penalty, cfg.contradiction_penalty)
    for e in ev["events"]:
        mem_store.add(e)

    return StressScenario(vol_mult=vol_mult, jump_mult=jump_mult, market=mk, belief=bl, event_world=ev, memory=mem_store)

# -----------------------------
# Precompute stress scenarios
# -----------------------------
t0 = _now()
local_rng = np.random.default_rng(CFG.seed + 999)

stress_points = [(vm, jm) for vm in CFG.stress_vol_mults for jm in CFG.stress_jump_mults]
scenarios: List[StressScenario] = []

print("Building stress scenarios:", len(stress_points))
for i, (vm, jm) in enumerate(stress_points, 1):
    if _now() - t0 > MAX_WALL_SEC * 0.45:   # reserve time for agent runs
        print("WARNING: wall-clock guard hit during scenario precompute; truncating grid.")
        break
    sc = build_scenario(CFG, local_rng, vm, jm)
    scenarios.append(sc)
    if i % 1 == 0:
        print(f"  scenario {i}/{len(stress_points)} done  (vol={vm}, jump={jm})")

log_event("stress_scenarios_built", {"count": len(scenarios), "grid_requested": len(stress_points)})

# -----------------------------
# Run agents over cached scenarios with progress + ETA
# -----------------------------
rows: List[Dict[str, Any]] = []
risk_flags: List[Dict[str, Any]] = []

total_jobs = len(scenarios) * len(agents)
job = 0
t_jobs = _now()
avg_job_sec = None

PARTIAL_PATH = AUDIT_DIR / "stress_results_partial.json"

for sc in scenarios:
    for ag in agents:
        job += 1
        if job > MAX_STRESS_RUNS:
            print("MAX_STRESS_RUNS reached; stopping.")
            break
        if _now() - t0 > MAX_WALL_SEC:
            print("MAX_WALL_SEC reached; stopping.")
            break

        t_job0 = _now()
        tr = run_agent(CFG, ag, sc.market, sc.belief, sc.memory, sc.event_world, local_rng)

        fe = float(tr.equity[-1])
        mdd = max_drawdown(tr.equity)
        es  = expected_shortfall(tr.pnl_inc, 0.95)
        mt  = float(np.mean(tr.turnover))
        fdd, feq, fch = compute_failure_flags(fe, mdd, mt)

        wall = float(_now() - t_job0)

        # Early-stop note (still record the result; just skip additional expensive extras if you add them later)
        catastrophic = (fe < EARLY_STOP_EQ) or (mdd > EARLY_STOP_DD)

        row = StressRow(sc.vol_mult, sc.jump_mult, ag.name, fe, mdd, es, mt, fdd, feq, fch, wall)
        rows.append(asdict(row))

        if fdd or feq or fch:
            risk_flags.append({
                "vol_mult": sc.vol_mult, "jump_mult": sc.jump_mult, "agent": ag.name,
                "flags": {"drawdown": fdd, "equity": feq, "churn": fch},
                "metrics": {"final_equity": fe, "mdd": mdd, "es_95": es, "mean_turnover": mt},
            })

        # progress + ETA
        if avg_job_sec is None:
            avg_job_sec = wall
        else:
            avg_job_sec = 0.92 * avg_job_sec + 0.08 * wall

        remaining = total_jobs - job
        eta = avg_job_sec * remaining

        # Print a heartbeat every few jobs so it never feels stuck
        if (job % 2 == 0) or catastrophic:
            msg = f"[{job}/{total_jobs}] vol={sc.vol_mult} jump={sc.jump_mult} agent={ag.name} "
            msg += f"eq={fe:.3f} mdd={mdd:.3f} turn={mt:.3f} job={wall:.2f}s ETA≈{_fmt_eta(eta)}"
            if catastrophic:
                msg += "  | catastrophic (recorded)"
            print(msg)

        # Incremental write (so you can interrupt safely)
        if job % 6 == 0:
            PARTIAL_PATH.write_text(json.dumps({
                "run_id": RUN_ID,
                "verification_status": "Not Verified",
                "partial": True,
                "rows": rows,
                "risk_flags": risk_flags[:500],
                "scenarios_built": len(scenarios),
                "jobs_done": job,
                "jobs_total": total_jobs,
            }, indent=2), encoding="utf-8")

    if (_now() - t0 > MAX_WALL_SEC) or (job > MAX_STRESS_RUNS):
        break

# Final artifact
stress = {"rows": rows, "risk_flags": risk_flags}
(AUDIT_DIR / "stress_results.json").write_text(
    json.dumps({"run_id": RUN_ID, "stress": stress, "verification_status": "Not Verified"}, indent=2),
    encoding="utf-8"
)

# Risk log update (bounded)
risk_log = {
    "run_id": RUN_ID,
    "timestamp_utc": utc_now_iso(),
    "verification_status": "Not Verified",
    "risks": [
        {"risk": "belief_misspecification", "trigger": "CUSUM alarms, posterior collapse, regime misalignment", "control": "mismatch alarms, resampling, model review"},
        {"risk": "context_interference", "trigger": "confidently_wrong_rate rises; gate_conf high + wrong", "control": "novelty + contradiction gating, deduping, decay"},
        {"risk": "hard_negative_dominance", "trigger": "retrieved_mass declines; evidence flips sign", "control": "provenance filters, secondary scoring, margin checks"},
        {"risk": "execution_debt", "trigger": "turnover spikes; convex impact dominates", "control": "turnover caps, vol targeting, stress suites"},
    ],
    "stress_failures": risk_flags[:200],
    "runtime": {"wall_sec": float(_now() - t0), "scenarios_built": len(scenarios), "jobs_done": job, "jobs_total": total_jobs},
}
(AUDIT_DIR / "risk_log.json").write_text(json.dumps(risk_log, indent=2), encoding="utf-8")
log_event("stress_done", {"flags": len(risk_flags), "rows": len(rows), "scenarios": len(scenarios)})

print("\nStress suite complete.")
print("  scenarios built:", len(scenarios))
print("  rows:", len(rows))
print("  flags:", len(risk_flags))
print("  wall_sec:", float(_now() - t0))

# Surface plot for A_full_governed (only for points we actually ran)
V = sorted({r["vol_mult"] for r in rows})
J = sorted({r["jump_mult"] for r in rows})
grid = {(r["vol_mult"], r["jump_mult"]): r["final_equity"]
        for r in rows if r["agent"] == "A_full_governed"}

Z = np.full((len(V), len(J)), np.nan, dtype=float)
for i, vm in enumerate(V):
    for j, jm in enumerate(J):
        if (vm, jm) in grid:
            Z[i, j] = grid[(vm, jm)]

plt.figure(figsize=(6.5,4.5))
plt.imshow(Z, aspect="auto")
plt.xticks(range(len(J)), [str(x) for x in J])
plt.yticks(range(len(V)), [str(x) for x in V])
plt.title("Stress Surface — A_full_governed Final Equity (cached scenarios)")
plt.xlabel("jump_mult")
plt.ylabel("vol_mult")
plt.colorbar()
plt.show()
```

##10.AUDIT BUNDLE

###10.1.OVERVIEW

**Cell 10 — Audit bundle finalization: hash ledger, zip packaging, and integrity self-validation**  
Cell 10 closes the chapter with integrity controls that make the entire run tamper-evident and portable. First, it enumerates all files in the audit directory and computes a SHA-256 hash for each, writing them into `hash_ledger.json`. This is the minimal cryptographic foundation for traceability. If any artifact is modified after creation—whether by accidental re-run, manual edit, or file corruption—the mismatch can be detected by recomputing hashes. In professional settings, this is essential: without integrity checks, governance artifacts can be quietly altered, undermining review trust.  

Second, the cell packages all artifacts into a compressed `audit_bundle.zip`. This is how the run becomes transferable evidence. A reviewer, auditor, or collaborator should be able to receive a single file that contains the manifest, logs, diagnostics, stress results, risk log, verification register, governance store, and hash ledger. Packaging is not clerical; it is the mechanism by which research becomes reviewable.  

Third, the cell performs self-validation. It opens the zip and checks that required core files exist. This prevents the failure mode you encountered earlier where a key governance file was missing. The cell then checks hash consistency for every artifact (excluding the zip itself). This turns integrity into an enforced invariant: the notebook will fail loudly if artifacts do not match the ledger, rather than silently producing a bundle with internal inconsistencies.  

Finally, the cell writes a `bundle_summary.json` that records counts, paths, timestamps, and verification status. This creates a final closure record and logs a “bundle packaged” event. The closure event is governance-relevant: it proves that the run completed its audit lifecycle.  

Conceptually, Cell 10 teaches that governance is operational. It is not a disclaimer at the top of the notebook. It is a chain of controls that begins with determinism and manifests, and ends with integrity-checked packaging. For long-context and temporal agents—especially in finance—this is the standard you want: outputs are not just produced; they are preserved, validated, and made review-ready.


###10.2.CODE AND IMPLEMENTATION

```python
# =============================================================================
# CELL 10 (FIXED v2) — Integrity Without Self-Hash Paradox
#   - hash_ledger.json hashes ALL artifacts EXCEPT: hash_ledger.json, seal.json, audit_bundle.zip
#   - seal.json hashes hash_ledger.json (and the zip optionally)
#   - verification checks ledger files + seal hash
# =============================================================================

def file_sha256(path: Path) -> str:
    h = hashlib.sha256()
    with path.open("rb") as f:
        for chunk in iter(lambda: f.read(1 << 20), b""):
            h.update(chunk)
    return h.hexdigest()

def list_audit_files(audit_dir: Path) -> List[Path]:
    return sorted([p for p in audit_dir.glob("*") if p.is_file()])

REQUIRED = {
    "run_manifest.json",
    "events_log.jsonl",
    "belief_trace_summary.json",
    "diagnostics.json",
    "stress_results.json",
    "risk_log.json",
    "verification_register.json",
    "diagnostics_bundle.json",
    "governance_store.json",
}

zip_path = AUDIT_DIR / "audit_bundle.zip"
bundle_summary_path = AUDIT_DIR / "bundle_summary.json"
hash_ledger_path = AUDIT_DIR / "hash_ledger.json"
seal_path = AUDIT_DIR / "seal.json"

# -----------------------------------------------------------------------------
# 0) Preconditions: required artifacts present
# -----------------------------------------------------------------------------
existing = list_audit_files(AUDIT_DIR)
existing_names = {p.name for p in existing}
missing_req = sorted([f for f in REQUIRED if f not in existing_names])
if missing_req:
    raise AssertionError(f"Missing required audit artifacts BEFORE bundling: {missing_req}")

# -----------------------------------------------------------------------------
# 1) Write bundle_summary ONCE (it will be included in ledger)
# -----------------------------------------------------------------------------
bundle_summary = {
    "run_id": RUN_ID,
    "timestamp_utc": utc_now_iso(),
    "audit_dir": str(AUDIT_DIR.resolve()),
    "required_files": sorted(list(REQUIRED)),
    "verification_status": "Not Verified",
    "integrity_protocol": "ledger_excludes_self; seal_hashes_ledger",
}
bundle_summary_path.write_text(json.dumps(bundle_summary, indent=2), encoding="utf-8")

# Clean old outputs
for p in [zip_path, hash_ledger_path, seal_path]:
    if p.exists():
        p.unlink()

# -----------------------------------------------------------------------------
# 2) Compute ledger EXCLUDING itself, seal, and zip
# -----------------------------------------------------------------------------
EXCLUDE = {"audit_bundle.zip", "hash_ledger.json", "seal.json"}

files_for_ledger = [p for p in list_audit_files(AUDIT_DIR) if p.name not in EXCLUDE]
hash_ledger = [{"file": p.name, "sha256": file_sha256(p)} for p in files_for_ledger]
hash_ledger_path.write_text(json.dumps(hash_ledger, indent=2), encoding="utf-8")

# -----------------------------------------------------------------------------
# 3) Create zip AFTER ledger exists (zip includes ledger + seal later)
# -----------------------------------------------------------------------------
with zipfile.ZipFile(zip_path, "w", compression=zipfile.ZIP_DEFLATED) as z:
    for p in list_audit_files(AUDIT_DIR):
        if p.name == "audit_bundle.zip":
            continue
        z.write(p, arcname=p.name)

# -----------------------------------------------------------------------------
# 4) Seal the bundle: hash the ledger (and zip hash optionally)
# -----------------------------------------------------------------------------
seal = {
    "run_id": RUN_ID,
    "timestamp_utc": utc_now_iso(),
    "verification_status": "Not Verified",
    "hash_ledger_sha256": file_sha256(hash_ledger_path),
    "audit_bundle_zip_sha256": file_sha256(zip_path),
}
seal_path.write_text(json.dumps(seal, indent=2), encoding="utf-8")

# IMPORTANT: zip currently does NOT include seal.json (it was created after zipping).
# Rebuild zip one more time to include seal.json without changing ledger-covered files.
with zipfile.ZipFile(zip_path, "w", compression=zipfile.ZIP_DEFLATED) as z:
    for p in list_audit_files(AUDIT_DIR):
        if p.name == "audit_bundle.zip":
            continue
        z.write(p, arcname=p.name)

# -----------------------------------------------------------------------------
# 5) Verification
#   (a) verify ledger-covered files match
#   (b) verify seal matches ledger hash
#   (c) verify zip contains required + bundle files
# -----------------------------------------------------------------------------
ledger_map = {r["file"]: r["sha256"] for r in json.loads(hash_ledger_path.read_text(encoding="utf-8"))}

for fname, expected_hash in ledger_map.items():
    fpath = AUDIT_DIR / fname
    if not fpath.exists():
        raise AssertionError(f"Ledger lists file that does not exist: {fname}")
    actual_hash = file_sha256(fpath)
    if actual_hash != expected_hash:
        raise AssertionError(f"Hash mismatch for {fname}: expected {expected_hash} got {actual_hash}")

seal_loaded = json.loads(seal_path.read_text(encoding="utf-8"))
ledger_hash_now = file_sha256(hash_ledger_path)
if seal_loaded["hash_ledger_sha256"] != ledger_hash_now:
    raise AssertionError("Seal mismatch: hash_ledger.json has changed since sealing")

with zipfile.ZipFile(zip_path, "r") as z:
    zipped = set(z.namelist())

must_have = REQUIRED | {"bundle_summary.json", "hash_ledger.json", "seal.json", "events_log.jsonl"}
missing_in_zip = sorted([f for f in must_have if f not in zipped])
if missing_in_zip:
    raise AssertionError(f"Zip missing required files: {missing_in_zip}")

final_summary = {
    "run_id": RUN_ID,
    "timestamp_utc": utc_now_iso(),
    "audit_dir": str(AUDIT_DIR.resolve()),
    "ledger_files_count": len(ledger_map),
    "total_files_count": len(list_audit_files(AUDIT_DIR)),
    "zip_path": str(zip_path.resolve()),
    "zip_contains": len(zipped),
    "verification_status": "Not Verified",
    "integrity": "PASS",
}

log_event("bundle_packaged", final_summary)

print("=== CH3 AUDIT BUNDLE READY (INTEGRITY PASS) ===")
print(json.dumps(final_summary, indent=2))
print("\nAudit directory contents:")
for p in sorted([x.name for x in AUDIT_DIR.glob("*")]):
    print(" -", p)
```

##11.CONCLUSION

**Conclusion — Temporal Intelligence as a Governed Control System: What Chapter 3 Proves and How to Use It**  
Chapter 3 resolves a deceptively simple question: what does it actually mean for an agent to be “temporal”? The popular answer is that a temporal agent can remember more. The chapter’s answer is stricter and more professional: a temporal agent is a **governed control system** that maintains uncertainty-aware beliefs over latent state, retrieves time-stamped evidence under interference constraints, converts belief and evidence into actions under feasibility limits, and produces a trace that can be audited end-to-end. If any one of those properties is missing, “memory” is not intelligence; it is accumulation. In finance terms, a system that stores history without controlled selection is like a trading desk that collects research without ever validating or reconciling it—eventually the volume of narratives overwhelms decision quality.

The first conclusion is that **state is not optional**. The chapter’s latent regime model is synthetic, but it captures the core fact that markets are nonstationary and partially observable. The agent does not observe regime directly; it observes returns drawn from regime-dependent distributions with heavy tails and jumps. A point-estimate classifier would be brittle here, because extreme moves and jump clusters are normal under the generator. The particle filter’s posterior distribution over regimes is the correct epistemic object: it expresses uncertainty and allows decisions to be conditioned on confidence. Effective sample size (ESS) becomes an operational diagnostic of belief health; resampling becomes an explicit mechanism to avoid degeneracy; and the posterior itself becomes a traceable state variable, not an internal black box. In real finance workflows, this maps to the difference between “the model says risk is high” and “the model assigns 70% probability to a high-volatility regime with heavy tails; uncertainty is elevated.” That difference determines whether you tighten risk budgets and request review or whether you trade as usual.

The second conclusion is that **model mismatch must be treated as a first-class risk**. The CUSUM residual alarms are not primarily forecasting tools; they are governance controls. A regime filter can be mathematically elegant yet wrong because the environment changes, because tails fatten beyond assumed levels, because jump intensity rises, or because correlation structure shifts. CUSUM creates an auditable record of sustained mismatch. In professional terms, it provides a trigger condition: a documented reason to shift into conservative mode, lower leverage, or freeze certain behaviors. The critical insight is that temporal intelligence is not only about being right; it is about knowing when you might be wrong and behaving accordingly.

The third conclusion is that **memory selection dominates memory storage**. The episodic event stream was designed to look like “context” rather than price data: embeddings, topics, polarity, and time stamps. The key result is that a large memory store is not helpful unless retrieval is governed by time, redundancy, and contradiction controls. Temporal decay is the baseline: relevance must be time-sensitive. Novelty gating addresses the repetition trap: redundancy in context can masquerade as confirmation, leading the agent to overweight duplicated claims. Contradiction gating stabilizes policy under noisy or adversarial evidence by requiring higher effective relevance for evidence that conflicts with belief-driven expectations. The chapter does not treat these as philosophical preferences; it treats them as tunable controls whose contribution is measured via ablation. This is the most important methodological advance over typical agent discussions: you do not “add memory” and hope; you design memory as a governed mechanism and validate what each gate prevents.

The fourth conclusion is that **hard negatives are the adversarial core of long context**. Chapter 2 showed that high-dimensional retrieval suffers from hubness and collisions. Chapter 3 shows the temporal form: in a time-stamped memory, hard negatives can remain persistently attractive to retrieval because they are aligned to the query direction, regardless of regime or time. Without kind penalties, novelty controls, and contradiction handling, hard negatives can dominate evidence, causing the agent to take high-confidence but wrong actions. This is the retrieval analogue of confident hallucination. The chapter operationalizes that risk through a “confidently wrong” metric: high gating confidence combined with directional wrongness relative to the true regime drift sign. This metric is more important than an average hit rate because it measures the dangerous case: decisive errors. In real finance, decisive errors are the ones that blow up risk budgets. An agent that is uncertain and wrong can be supervised; an agent that is confident and wrong is a liability.

The fifth conclusion is that **execution realism turns temporal reasoning into a viability test**. The agent does not merely predict; it acts. It trades under volatility targeting, leverage caps, and turnover limits, and it pays spreads, fees, and convex impact. This transforms the memory-belief system from a classifier into a closed-loop policy. In quiet regimes, an agent can churn without consequence in unrealistic models. In this lab, churn produces execution debt; impact costs grow nonlinearly with turnover; spreads widen under stress and jumps. This mirrors the real world: during turbulence, both inference and execution get harder. The professional takeaway is that temporal intelligence must be evaluated as a system: belief inference that improves directional alignment but increases turnover can still be net harmful if it pushes the strategy onto a punitive execution surface. Conversely, a conservative gating policy might lower raw responsiveness but improve survival by reducing churn and convex cost exposure.

The sixth conclusion is methodological: **ablation is governance**. The chapter’s agent suite does not exist to create a leaderboard. It exists to isolate causal mechanisms. Removing decay tests whether staleness control matters. Removing novelty tests whether redundancy control matters. Removing contradiction tests whether conflict handling reduces whipsaw. Removing retrieval tests whether episodic memory is actually adding value beyond belief. Removing belief tests whether memory can substitute for latent state inference. This is how you create teachable, defensible conclusions. In a professional setting, ablation provides the language of controls: you can say “this gate reduces churn under stress,” “this filter reduces confident wrongness,” and “this belief state improves alignment but requires alarm gating.” That is far more valuable than saying “the model performs well,” which is neither diagnostic nor governable.

The seventh conclusion is that **robustness must be mapped as a surface, not asserted as a claim**. The stress suite reruns the full pipeline under elevated volatility and jump intensity. It produces failure flags that are explicit and machine-readable. This is how a risk function becomes operational: you define what failure means (drawdown, equity collapse, churn), you test across structural stress states, and you record where failures occur. The stress surface plot is not a marketing figure; it is a map of fragility. It answers the only question that matters in regulated or high-stakes environments: under what conditions does this system break, and how do we know? In finance, this is the difference between a backtest and a risk program.

The eighth conclusion is governance-first in the strictest sense: **traceability is not optional; it is the deliverable**. The chapter produces a run manifest, append-only event logs, diagnostics artifacts, stress results, risk logs, verification registers, a consolidated governance store, hash ledgers, and a self-validated zip bundle. This is the professional standard for any system that influences decisions. It allows a reviewer to reconstruct the run without trusting the notebook author. It makes silent edits detectable. It makes assumptions and limitations explicit. It creates a disciplined separation between what is demonstrated and what is unverified. In finance practice, this aligns with model risk management: a model is not only its outputs, but the evidence that its outputs were produced under known conditions and that its risks are documented.

So what does Chapter 3 ultimately prove? It proves that temporal intelligence is not “long context.” It is a structured and governed pipeline: latent belief, controlled memory retrieval, constrained policy execution, and auditable artifacts. It also proves that increasing capability increases risk. More memory increases interference. More responsiveness increases churn and execution debt. More confidence increases the danger of wrongness when adversaries exist. Therefore controls must scale with capability: decay, novelty, contradiction gates; alarms and posterior diagnostics; turnover and leverage caps; stress surfaces; audit bundling. The chapter’s most important contribution is that it turns these controls into concrete code, metrics, and artifacts rather than opinions.

If you want to use this chapter as a teaching module for finance practitioners, the correct framing is not “here is an agent that trades.” The correct framing is “here is how to build a temporally aware decision system that remains reviewable under stress.” The goal is to teach mechanism literacy: how belief inference behaves under heavy tails, how long context becomes adversarial, how gating reduces interference, how execution turns signals into costs, and how governance turns experiments into evidence. That is the real PhD-level lesson: intelligence is not the absence of error. It is the disciplined management of error, uncertainty, and interference—under controls strong enough to be audited.
